#### Defining the function

Now, we'll see the problem of $ y(\boldsymbol{x},\boldsymbol{w}) = w_0 + w_1 x^1 + ... + w_D x^D $ and make some changes. Let's choose some arbitrary function class $ \phi(x) $ and then, our previous relation becomes $ y(\boldsymbol{x},\boldsymbol{\phi}) = w_0 + w_1 \phi_1 + ... + w_D \phi_D $, such that $ \phi_i $ are apparently linearly independent. So, in this example we'll choose the function $$ \phi_i = \exp \left( -\frac{(x-\mu_i)^2}{s^2} \right) $$

```python
def gauss(x,mu,s):
    import numpy as np
    phi = np.e**(-((x-mu)**2) / (s**2))  
    
    return phi
```

#### Setting samples

Let's choose too the means values in which we'll fit the distribution. In example, we'll make

```python
def linear_regressor(x_data, y_data, M, pseudo=True):
    M = M+1
    
    xsum = np.zeros((1,M*2))
    

    for i in range(M*2):
        mu = i / len(y_data)
        xsum[0][i] = sum(gauss(x_data, mu, 1))
    
    S = np.zeros((M,M))
    Y = np.zeros((1,M))

    for i in range(M):
        for j in range(M):
            S[i][j] = xsum[0][i+j]
        Y[0][i] = sum(y_data*(x_data ** i))
    
    # S * W = Y

    if pseudo == False:
        W = np.linalg.solve(S,Y.T)
        W = W.T
    else:
        S_aux = np.dot(S,S.T) # S * S.T
        S_aux = np.linalg.inv(S_aux)
        S_inv = np.dot(S_aux,S.T) # S_aux * S.T
        W = np.dot(Y,S_inv) # Y * S_inv
    
    y_pred = 0
    
    x = np.linspace(x_data[0],x_data[len(x_data)-1],50)

    for i in range(M):
        y_pred = W[0][i]*(x**i) + y_pred
    
    return y_pred
```

```python
y_pred_test = linear_regressor(x,data,3)
```
