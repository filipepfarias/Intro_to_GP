```python
ns = int(1e2)

x_lim = np.linspace(0,1,ns)
y_lim = x_lim

X,Y = np.array(np.meshgrid(x_lim,y_lim))
```

```python
Z = np.sin(2*np.pi*X)*np.sin(2*np.pi*Y)
T = Z + np.random.normal(0,0.1,Z.shape)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Plot the surface.
surf = ax.plot_surface(X, Y, T)

plt.show()
```
